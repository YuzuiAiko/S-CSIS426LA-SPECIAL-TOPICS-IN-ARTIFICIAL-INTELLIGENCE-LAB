```python
# Step 0: Install required packages
!pip install tensorflow scikit-learn pandas

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
import numpy as np
```

```python
# -----------------------------
# Step 1: Create a sample dataset
# -----------------------------
emails = [
    "Meeting schedule for academic project",
    "Your account password reset request",
    "Exclusive promo: Buy 1 take 1 coffee",
    "Urgent: Verify your bank account now",
    "Invitation to research seminar",
    "Congratulations! You won a free iPhone",
    "Reminder: Submit your thesis draft",
    "Limited time offer on new gadgets"
]

categories = [
    "Important",      # Academic
    "Important",      # Personal
    "Promotional",
    "Phishing",
    "Important",      # Academic
    "Phishing",
    "Important",      # Academic
    "Promotional"
]

df = pd.DataFrame({"Email": emails, "Category": categories})

# Encode categories numerically
label_map = {"Important":0, "Promotional":1, "Phishing":2}
df["Label"] = df["Category"].map(label_map)

# -----------------------------
# Step 2: Vectorize the emails
# -----------------------------
vectorizer = TfidfVectorizer(stop_words="english")
X = vectorizer.fit_transform(df["Email"]).toarray()
y = to_categorical(df["Label"], num_classes=3)

# -----------------------------
# Step 3: Train/Test Split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# -----------------------------
# Step 4: Build Neural Network
# -----------------------------
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X.shape[1],)))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3 categories: Important, Promotional, Phishing

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# -----------------------------
# Step 5: Train the Model
# -----------------------------
model.fit(X_train, y_train, epochs=10, batch_size=2, verbose=1)

# -----------------------------
# Step 6: Make Recommendations
# -----------------------------
def recommend_important_emails(query, top_n=3):
    query_vec = vectorizer.transform([query]).toarray()
    prediction = model.predict(query_vec)
    predicted_class = np.argmax(prediction)

    # Only recommend if predicted as Important
    if predicted_class == 0:
        # Find similar Important emails
        important_emails = df[df["Category"]=="Important"]["Email"].tolist()
        return important_emails[:top_n]
    else:
        return ["No Important recommendations found."]

# -----------------------------
# Step 7: Test the Recommendation Engine
# -----------------------------
user_query = "Looking for academic project updates"
print("User Query:", user_query)
print("\nRecommended Important Emails:\n")
print(recommend_important_emails(user_query))
```

```python
This notebook demonstrates an email classification and recommendation system using a simple neural network. Here's a breakdown of the code's functionality:

1.  **Setup and Imports**: Installs and imports necessary libraries like `tensorflow`, `scikit-learn`, and `pandas` for data handling, text processing, and deep learning.
2.  **Sample Data Creation**: Creates a small dataset of emails, each assigned a category: "Important", "Promotional", or "Phishing". These categories are then numerically encoded.
3.  **Text Vectorization**: Uses `TfidfVectorizer` to convert the email text into numerical feature vectors, making them suitable for machine learning models.
4.  **Data Splitting**: Divides the dataset into training and testing sets to evaluate the model's performance.
5.  **Neural Network Definition**: Constructs a simple Sequential neural network using Keras, consisting of `Dense` layers and `Dropout` for regularization. The network is configured to classify emails into 3 categories.
6.  **Model Training**: Trains the defined neural network using the prepared training data.
7.  **Recommendation Function**: Implements a function `recommend_important_emails` that takes a user query, predicts its category using the trained model, and if the query is classified as "Important", it returns a list of existing "Important" emails from the dataset.
8.  **Testing**: Demonstrates how to use the `recommend_important_emails` function with a sample query and prints the recommendations.
```
