**VACUUM CLEANER AGENT (EXAMPLE)**

```python
def vacuum_agent(location, status):
    if status == "Dirty":
        return "Suck"
    elif location == "A":
        return "Move Right"
    elif location == "B":
        return "Move Left"

# Simulation
test_data = [("A", "Dirty"), ("A", "Clean"),
             ("B", "Dirty"), ("B", "Clean")]

for location, status in test_data:
    action = vacuum_agent(location, status)
    print("Location:", location, "| Status:", status, "| Action:", action)
```

**IMPLEMENTATION - VERSION 1 (ON & OFF WITH REFLEX)**

```python
def temperature_agent_basic(temp_celsius):
    """
    Simple reflex agent:
    - If temperature > 28°C => Fan ON
    - Otherwise => Fan OFF
    """
    if temp_celsius > 28:
        return "Fan ON"
    return "Fan OFF"

# Example simulation
temperatures = [30, 25, 29, 27, 31, 26]
for t in temperatures:
    print(f"Temp: {t} °C -> {temperature_agent_basic(t)}")
```

**IMPROVEMENTS - VERSION 2 (HYSTERESIS + ADJUSTATBLE SPEED + TIMER)**

```python
import time
class SmartFanAgent:
    def __init__(self, lower=26, upper=29):
        self.lower = lower
        self.upper = upper
        self.state = "OFF"

    def decide(self, temp):
        if temp >= self.upper:
            self.state = "HIGH"
        elif temp > self.lower:
            self.state = "LOW"
        else:
            self.state = "OFF"
        return self.state


# Create agent
agent = SmartFanAgent()

# Test temperatures
temperatures = [25, 27, 28, 30, 29, 26, 24]

print("=== Smart Fan Simulation ===")
for t in temperatures:
    action = agent.decide(t)
    print(f"Temperature: {t}°C -> Fan State: {action}")

class SmartFanAgent:
    def __init__(self, lower=26, upper=29, cooldown_seconds=120):
        """
        lower: temperature (°C) below which we ensure fan is OFF.
        upper: temperature (°C) above which we ensure fan is ON.
        cooldown_seconds: minimum seconds between mode changes to avoid rapid toggling.
        """
        self.lower = lower
        self.upper = upper
        self.state = "OFF"          # current actuator state: OFF, LOW, MED, HIGH
        self.last_change = 0        # timestamp of last change
        self.cooldown = cooldown_seconds

    def decide(self, temp):
        now = time.time()
        # Cooldown logic to prevent frequent switching:
        if now - self.last_change < self.cooldown:
            return self.state

        # Hysteresis-based decision:
        if temp >= self.upper:
            new_state = "HIGH"
        elif temp >= (self.lower + (self.upper - self.lower)/2):
            new_state = "MED"
        elif temp > self.lower:
            new_state = "LOW"
        else:
            new_state = "OFF"

        if new_state != self.state:
            self.state = new_state
            self.last_change = now
        return self.state
```
